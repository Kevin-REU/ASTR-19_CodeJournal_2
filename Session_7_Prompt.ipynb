##### Create a Jupyter notebook, import matplotlib. Write cells that create an array x ranging from [0,1] in 100 steps and that defines a function that returns exp(x).  In a new cell use the function to set y=exp(x), and then plot x vs. y. Label the x-axis as “Time [milliseconds]” and the y-axis as “Awesomeness”. Save the figure as a PDF.

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
```

```python
def exp(x):
    return np.exp(x)                    # Calculate y = e^(x)
```

```python
def plot(x,y):
    fig = plt.figure(figsize=(10,10))   # Create a 10x10 inch image
    ax = fig.add_subplot(111)           # Create a single subplot
    plt.plot(x,y,'r-',label = 'y')      # Plot the values using a solid red line and label it 'y'
    plt.legend(loc = 2)                 # Add a legend on the upper left of the image
    plt.xlabel('Time [milliseconds]')   # Add an x label
    plt.ylabel('Awesomeness')           # Add a y label  
```

```python
def save():
    plt.savefig("ASTR-19_TimeInClass-to-Awesomeness.pdf",bbox_inches="tight",facecolor="white") # Save the plot
```

```python
def main():
    x = np.linspace(0,1,100)            # x = [0,1] in 100 steps
    y = exp(x)                          # Call for exp(x): y = e^(x) for all 100 steps
    plot(x,y)
    save()
```

```python
if __name__=="__main__":
    main()
```
