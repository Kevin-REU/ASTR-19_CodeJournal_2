##### Create a Jupyter notebook, import matplotlib. Write cells that create an array x ranging from [0,1] in 100 steps and that defines functions that return sin(x) and cos(x).  In a new cell use to create a multipanel plot (1 row, 2 columns), plotting sin(x) vs. x in the left panel and cos(x) vs. x in the right panel. Label the panels with sin(x) and cos(x), and save the figure as a PDF.

### Import Libraries

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

### Functions

```python
def xarray():
    return np.linspace(0,1,100)
```

```python
# Calculate the value of sin(x)
def sin(x):
    return np.sin(x)
```

```python
# Calculate the calue of cos(x)
def cos(x):
    return np.cos(x)
```

```python
def DualPanelPlot(x,                          # x varible
                  flag_save=True,             # Save the figure? -> Yes
                  xlabel='x',                 # x-axis label
                  lcolor='blue',              # Line color
                  fs=14,                      # Font size
                  fname='DualPanelPlot.pdf'): # Default plote filename

    # By default, flag_save = True, and we'll save the figure to a file. The plot fname = 'plot.png' by default, but we can change this.

    # Define a figure and axis
    nrow = 1
    ncolumn = 2
    f, ax = plt.subplots(nrow,ncolumn,figsize=(5*ncolumn,5*nrow))
    

    # Set the left panel
    y = sin(x)
    # Plot sin(x) vs. x
    ax[0].plot(x,y,color=lcolor,linewidth=1.5,label='model (sin(x))') 
    # Label the axes
    ax[0].set_xlabel(xlabel,fontsize=fs)
    ax[0].set_ylabel('sin(x)',fontsize=fs)

    # Set the right panel
    y = cos(x)
    # Plot cos(x) vs. x
    ax[1].plot(x,y,color=lcolor,linewidth=1.5,label='model (cos(x))')
    # Label our axes
    ax[1].set_xlabel(xlabel,fontsize=fs)
    ax[1].set_ylabel('cos(x)',fontsize=fs)
    

    # Fit the layout
    plt.tight_layout()
    # Save the Plot?
    if(flag_save):
        plt.savefig(fname,bbox_inches='tight',dpi=600)
```

### The Main function

```python
def main():
    # Yippee!!!
    print('Making two plots!!! :O')

    # Make an array of variables
    x = xarray()

    # Make the plot
    DualPanelPlot(x)
```

### Execute main()

```python
if __name__=='__main__':
    main()
```
